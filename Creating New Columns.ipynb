```python
import pandas as pd
df = pd.read_csv('nba.csv')
```

```python
df['Scorer'] = ""

for i in range(0,len(df)):
    if df.PTS[i]>2000:
        df.Scorer[i] = 'High Scorer'
    elif df.PTS[i]>500:
        df.Scorer[i] = 'Medium Scorer'
    else:
        df.Scorer[i] = 'Low Scorer'
        
df
```
