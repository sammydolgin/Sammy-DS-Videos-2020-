```python
import pandas as pd
pd.read_csv('nba.csv')
```

# One entry

```python
my_dictionary = {'name':['sammy'],'gender':['male'],'favorite_color':['green']}
pd.DataFrame.from_dict(my_dictionary)
```

# Multiple Entries

```python
my_dictionary = {'name':['sammy','alex','richard','katie'],'gender':['male','male','male','female'],'favorite_color':['green','blue','red','blue']}
pd.DataFrame.from_dict(my_dictionary)
```

# .keys(), .values(), .items()

```python
my_dictionary.keys()
my_dictionary.values()
my_dictionary.items()
```
