```python
#Input Variable (dataframe)
Input = data1[['petal width (cm)']]

#Output Variable (series)
Output = data1['sepal length (cm)']

#Defining the model
lm = linear_model.LinearRegression()

#Fit the model
lm.fit(Input, Output)

#Print the intercept
print('Intercept: ' + str(lm.intercept_))

#Print the coefficient
print('Coefficient ' + str(lm.coef_))

print('\n')

#Print the full model
print('Sepal Length (CM) = ' + str(round(lm.intercept_,3)) + ' + ' + str(round(lm.coef_[0],3)) + '*(Petal Width (CM))')
```

```python
#Select first 3 columns
input_vars = list(data1.columns[0:3])

#Input (to be a dataframe structure)
Input = data1[input_vars]

#Output (to be a series structure)
Output = data1['petal width (cm)']

#Define the model type
lm = linear_model.LinearRegression()

#Fit model
lm.fit(Input, Output)

#Print Intercept
print(lm.intercept_)

#Print Coefficient
print(lm.coef_)
```

Petal Width = -0.240 + -0.207*(Sepal Length) + 0.222*(Sepal Width) + 0.524*(Petal Length)

# Model Accuracy

```python
lm.score(Input,Output)
```

# Train/Test Split

```python
# Fitting the models

X_train, X_test, y_train, y_test = train_test_split(Input, Output, test_size=0.2, random_state=0)
lm.fit(X_train, y_train) #fit the model
print(lm.score(X_train, y_train)) #how it performs on the data it was trained on
```
